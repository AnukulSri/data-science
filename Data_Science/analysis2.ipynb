**missing data situation**

- [Case] in a column with numeric values
    - [fix] fill all the values by the mean of the column
- [Case] in a cloumn with non-numeric values/categorial values # categorial means repeating values
    - [fix] fill the values by the most frequent value of that columns
- [Case] in a cloumn with critical data
    - [fix] drop the records with missing data 

```python
pd.set_option('display.max_rows',None)
cs[cols_to_fix]
```

```python
mean_nl = cs['normalized-losses'].astype('float').mean()
cs['normalized-losses'].replace(np.nan,mean_nl,inplace=True) # np.nan means empty sapce
cs
```

```python
mean_hs=cs['horsepower'].astype('float').mean()
cs['horsepower'].replace(np.nan,mean_hs,inplace=True)
cs
```

```python
mean_bore=cs['bore'].astype('float').mean()
cs['bore'].replace(np.nan,mean_bore,inplace=True)
cs
```

```python
mean_stroke=cs['stroke'].astype('float').mean()
cs['stroke'].replace(np.nan,mean_stroke,inplace=True)
cs
```

```python
print(cs['peak-rpm'])
```

```python
mean_rpm=cs['peak-rpm'].astype('float').mean()
cs['peak-rpm'].replace(np.nan,mean_rpm,inplace=True)
cs
```

```python
mf_doors = cs['num-of-doors'].value_counts().idxmax()
cs['num-of-doors'].replace(np.nan,mf_doors,inplace=True)
cs
```

```python
cs.dropna(inplace=True) # here the cloumns which have empty sapce will be droped
cs
```

```python
cs['price']= cs['price'].astype('float')
cs['normalized-losses']= cs['normalized-losses'].astype('int')
cs['horsepower']= cs['horsepower'].astype('int')
cs['peak-rpm']= cs['peak-rpm'].astype('int')
cs['bore']= cs['bore'].astype('float')
cs['stroke']= cs['stroke'].astype('float')
```
