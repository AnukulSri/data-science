**actual processing for this dataset**

- drop the columns that are not needed
- rename the columns
- rename country with very long name(optional)
- convert all the columns to str name
- store the range of years in to a variable
- calculate

```python
df.head()
```

```python
cols_to_drop = ['Type','AREA','Coverage','REG','DEV'] # this is to drop these columns
df.drop(columns= cols_to_drop, inplace=True) # this means to drop the data in col_to_drop variable and the save it. inplace is used to save the changes after dropping
```

```python
df.head(4)
```

```python
# to rename the columns name:

df.rename({
           'OdName':'Country',
           'AreaName':'Continent',
           'RegName' : 'Region',
           'DevName': 'Status',
}, axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns=df.columns.astype(str) # this is to convert the column name in sting which is in the numeric form
```

```python
years=list(map(str,range(1980,2014)))
print(years)
```

```python
df[years]
```

```python
df['Total'] = df[years].sum(axis=1) # axis is counting the total no. of imigration of each columns seprately
df.head(3)
```

```python
plt.style.use('seaborn')
```

```python
df.set_index('Country',inplace=True)
df.head()
```

```python
continent_df = df.groupby('Continent').sum() # grouping data by continent
```

```python
filter_europe = df['Continent'] == 'Europe' # it will print the country of europe
eu_df = df[filter_europe].copy()
```

```python
#eu_df.columns[continent_df]
eu_df.sort_values(by='Total',ascending=False, inplace=True)
eu_df.head()[years].T
```

```python
eu_df.head()[years].T.plot(kind='line',
                          figsize=(20,5),
                          color=['green','red','blue','yellow','black'],
                          linewidth=3,
                          marker='o',
                          title = 'Immigration From Europe',
                          markersize=7
                          )
```

```python
eu_df.head()[years].T.plot(kind='line',
                          figsize=(20,5),
                          color=['green','red','blue','yellow','black'],
                          linewidth=3,
                          marker='o',
                          title = 'Immigration From Europe',
                          markersize=3,
                          subplots=True,
                          layout = (2,3))
```

```python
eu_df.sort_values(by='Total', ascending=False, inplace=True)
eu_df.head(6)[years].T.plot(kind='line',
                            figsize=(20,6),
                            linewidth=2,
                            marker='o',
                            markersize=5,
                            title='Top countries with immigration in Europe',
                            subplots=True,
                            layout = (2,3)
                            )
plt.show()
```

```python
df.sort_values(by='Total',ascending=False,inplace=True)
df.head(6)[years].T.plot(kind='area',
                            figsize=(20,6),
                            linewidth=2,
                            title='Top countries with immigration in World',
                            subplots=True,
                            layout = (2,3)
                            )
plt.show()
```

```python
df.sort_values(by = 'Total',ascending = False,inplace=True)
df.head(6)[years].T.plot(kind = 'area',
                        figsize=(20,5),
                        linewidth =2,
                        title = 'Top countries immigration from World',
                        stacked= False,
                        fontsize=12
                        )
plt.show()
```
