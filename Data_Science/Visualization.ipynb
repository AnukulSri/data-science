```python

import numpy as np
import matplotlib.pyplot as plt # visualization as image
#import seaborn as sns # statistical data visualization
import plotly.express as px # interactive plotting

a= np.random.randint(1,100,size=50)
print(a)
```

```python
plt.style.available # it shows the color vailable to change the color of graph
```

```python
plt.style.use('seaborn-deep')  # it is used to change the color of graph
```

```python
plt.figure(figsize=(25,5),facecolor='pink') # use to chnage the width of a graph. Facecolor is used to change the color of background
plt.plot(a) # it means to place the point on the graph
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(a,
color ='red', # color of line 
linewidth=2, # line width
linestyle=':', # which type of line to draw on graph
#marker='*',
markersize=10 
)

plt.title('Simple Line Plot demo',size = 20) #title is used to add the title on the graph
plt.xlabel('X-axis data',size = 20) # it is used to add the content on x axis like as title
plt.ylabel('Y-axis data',size = 20) # it is used to add the content on the y-axis 
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(a,'g--o')# here g is used for drawing line in green color and o is used as marker
plt.xticks(np.arange(0,51,2))
plt.show()
```

```python
fig = plt.figure(figsize=(25,5))
plt.plot(a,'g--x')
name =['A','B','C','D','E','F','G','H','I','J','K']
plt.xticks([1,5,10,15,20,25,30,35,40,45,50],labels=name)
plt.show()
```

```python
data1 = np.random.randint(0,100,50) # creating another variable with no. b/w 1 to 100 and no. should be 
data2 = np.random.randint(0,100,50)
fig = plt.figure(figsize=(25,5))
plt.plot(data1,'r:',label='data1')
plt.plot(data2,label='data2')
plt.legend(loc='lower right',fontsize=16)
#plt.show()
plt.savefig('sample.png',)
```

```python
plt.figure(figsize=(25,10))
plt.subplot(2,1,1)
plt.plot(data1,'r:',label='data1')
plt.legend(loc='upper right', fontsize=16)
plt.subplot(2,1,2)
plt.plot(data2,label='data2')
plt.legend(loc='lower right',fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25,10),nrows=2,ncols=1,sharex=True) # 0 and 1 are index of subplot
ax[0].plot(data1,'rx:',label='data1',)
ax[1].plot(data2,'bo--',label='data2')
ax[0].legend(loc='lower right',fontsize=16)
ax[1].legend(loc='lower right',fontsize=16)
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```
