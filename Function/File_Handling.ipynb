# File Handling

In coding we have 3 types of files:
- text file(.txt, .py, .css, .js, etc)
- binary files (.jpg, .png,.zip, .mp3, .mp4, etc)
- proprietry files (.docx, .xls, .pptx, .pdf, etc)

file handling is the process of:
- opening 
- reading 
- modifying and 
- closing files.

# Reading A File

```python
f= open('Turtle.py') # here f is a variable in which we are opening a file. To open a file there is open function in python.
c= f.read() # Now here c is an another variable in which we are reading a file or we can say we are storing files data. To read file there is a function read().
f.close() # after reading a file we have to close that file, so to close a file there is close function.
print(c) 
```

# Writing content in a Text File

```python
story = 'once upon a time there is a planet that was green, and blur, the end'
file = open('writing_Content.txt','w')  # w is to write a content in a file
file.write(story)                 
file.close()
```

```python
info = input('enter the info:')
with open('writing_text2.txt','w') as f: # here w is used to write the content in a file
    f.write(info)
    f.write('\n')
    f.write('author\n')
    f.write('Anukul Srivastava')
```

# Adding content to a file
